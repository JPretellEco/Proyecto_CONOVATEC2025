# Lectura del parquet

```python
# Leemos el archivo parquet 

import pandas as pd
dataset = pd.read_parquet('/home/jefferssonds/Documentos/datathon/data/casos_piura_limpio.parquet')
data = dataset.copy()
```

# Identificamos qué tipo de fechas tenemos

```python
print('Rango de fecha de ingreso del registro al sistema',data['FECHA_ING'].min(), '-',data['FECHA_ING'].max())

print('-------------------------FECHAS DE EXTRACCIÓN------------------------------------')
print('Una sola fecha de recogió la muestra de la información',data['FECHA_MUESTRA'].max())
print('Una sola fecha de corte de la recolección de información',data['FECHA_CORTE'].max())
```

```python
data.groupby('ANO')[['DAA_C1', 'DAA_C1_4', 'DAA_C5']].agg(['sum', 'mean', 'max'])
```

```python
data.loc[data['DAA_C5'] == 0, 'DAA_C5'] = (
    data.loc[data['DAA_C5'] == 0, 'DAA_C1'] + data.loc[data['DAA_C5'] == 0, 'DAA_C1_4']
)
```

```python
import datetime

data['FECHA_SEMANA'] = data.apply(
    lambda row: pd.to_datetime(datetime.date.fromisocalendar(int(row['ANO']), int(row['SEMANA']), 1)),
    axis=1
)
```

```python
data.rename(columns={'FECHA_SEMANA': 'FECHA_SINTOMAS'}, inplace=True)
```
